### Defining dehumanization binary

```python
import re
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk


nltk.download('punkt')
stemmer = PorterStemmer()
```

```python
animal = {stemmer.stem(word) for word in ['parasites', 'leeches', 'bloodsucker']}
natural = {stemmer.stem(word) for word in ['flood', 'influx', 'wave', 'surge', 'tide', 'massive', 'flow', 'avalanche', 'contingent', 'overflow', 'overcrowding', 'invasion', 'drown', 'pouring']}
illegal = {stemmer.stem(word) for word in ['criminal', 'illegal', 'unlawful']}
economic = {stemmer.stem(word) for word in ['economic', 'money', 'fund', 'cost', 'funding', 'expense', 'pay', 'fiscal','price', 'burden', 'investment', 'allowance', 'collect', 'bank', 'donation', 'income', 'capita', 'buy', 'deficit', 'tax', 'market', 'budget', 'credit', 'account', 'debt', 'dollar', 'deposit', 'incentive', 'rent', 'unemployed', 'unskilled']}
object = {stemmer.stem(word) for word in ['boats', 'allocate', 'distribute', 'redistribute', 'dock', 'stranded', 'stuck', 'share', 'send', 'transfer', 'anchored', 'assume', 'deport', 'camp', 'smuggle']}
numbers = {stemmer.stem(word) for word in ['thousands', 'millions', 'quota', 'hundreds', 'tens', 'increase', 'tons', 'quota', 'increasing', 'many', 'double', 'numerous', 'dozens', 'multitude', 'triple']}
```

```python
def count_keywords(s, keywords):
    words = word_tokenize(s.lower())  # Tokenize and handle case insensitivity
    stemmed_words = [stemmer.stem(word) for word in words]  # Stem each word
    return sum(1 for word in stemmed_words if word in keywords)
```

```python
# Applying the function to each dictionary and each row in 'SVO' column
svo_clusters['animal_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, animal))
svo_clusters['natural_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, natural))
svo_clusters['illegal_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, illegal))
svo_clusters['economic_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, economic))
svo_clusters['object_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, object))
svo_clusters['numbers_count'] = svo_clusters['string_SVO_col'].apply(lambda s: count_keywords(s, numbers))
```

```python
svo_clusters.columns
```

```python
svo_clusters['total_dictionaries'] = svo_clusters['animal_count'] + svo_clusters['natural_count'] + svo_clusters['illegal_count'] + svo_clusters['economic_count'] + svo_clusters['object_count'] + svo_clusters['numbers_count']
```

#### Combining Results

```python
def calculate_aggregations(df):
    # Group by 'hdb_string_clusters'
    grouped = df.groupby('hdb_string_clusters')
    # Calculate outlet proportions
    outlet_counts = grouped['Outlet'].value_counts().unstack(fill_value=0)
    outlet_proportions = outlet_counts.div(outlet_counts.sum(axis=1), axis=0).round(2)
    means = grouped[['Perspective(ws)', 'Perspective(wo)', 'Value(o)', 'Value(s)']].mean().round(2)
    # Calculate sums for other specified columns
    sums = grouped[['EU_Mentions', 'NonEU_Mentions', 'GN_Mentions', 'GS_Mentions', 'Total_Count', 'n_articles', 'natural_count', 'illegal_count', 'economic_count', 'object_count', 'numbers_count', 'total_dictionaries']].sum()
    # Combine the results into a single DataFrame
    combined = outlet_proportions.join(means).join(sums)
    return combined


def get_top_svos(df, n_top=5):
    # Group by 'hdb_string_clusters' and 'string_SVO', then count occurrences
    freq = df.groupby(['hdb_string_clusters', 'string_SVO_col']).size().reset_index(name='counts')
    # Sort within each cluster by count in descending order and get top N
    top_svos = freq.groupby('hdb_string_clusters').apply(
        lambda x: '\n'.join(x.sort_values('counts', ascending=False).head(n_top)['string_SVO_col'])
    ).reset_index(name='Top_SVOs')
    return top_svos

aggregated_data = calculate_aggregations(svo_clusters)
top_svos = get_top_svos(svo_clusters)
result_df_dehuman= top_svos.merge(aggregated_data, on='hdb_string_clusters', how='left')
```
